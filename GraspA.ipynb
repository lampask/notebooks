Firstly we need to get initial random data from dice rolls. The code in this part rolls two dice 50 times and then shows the resulting pairs below:

```python
import random
a = [(random.randrange(1,7), random.randrange(1,7)) for i in range(50)]
print(a)
```

The next step is to process the dice numbers into locations on the model of atom. Each number here is representing the resulting layer for the pair of dice rolls:

```python
b = [((i[0]+i[1])/2) for i in a]
print(b)
```

Lastly we need to show our results in a table and a nice atom model:

```python
import numpy as np
import matplotlib.pyplot as plt
from random import randrange
import pandas as pd
import math

theta = np.linspace(0, 2*np.pi, 100)
fig, ax = plt.subplots(1)

for i in range(7):
    ax.plot(np.sqrt(0.1*i)*np.cos(theta), np.sqrt(0.1*i)*np.sin(theta))
    ax.plot(0, 0, "ro", markersize=20)

ax.set_aspect(1)
for i in b:
    l = randrange(0,360)
    ax.plot(math.sqrt(0.1*i)*math.cos(l), math.sqrt(0.1*i)*math.sin(l), "ro", markersize=2, mec='black', mfc='black')
    
plt.xlim(-1.25,1.25)
plt.ylim(-1.25,1.25)
plt.grid(linestyle='--')
plt.title('MODELLING THE ELECTRON LOCATION IN A HYDROGEN ATOM', fontsize=8)
plt.savefig("grasp.png", bbox_inches='tight')

plt.show()
data = [['█' if i == 1.0 else '',
        '█' if i == 1.5 else '',
        '█' if i == 2.0 else '',
        '█' if i == 2.5 else '', 
        '█' if i == 3.0 else '', 
        '█' if i == 3.5 else '', 
        '█' if i == 4.0 else '', 
        '█' if i == 4.5 else '', 
        '█' if i == 5.0 else '', 
        '█' if i == 5.5 else '', 
        '█' if i == 6.0 else ''] for i in b]
pd.DataFrame(data, columns=["1.0","1.5","2.0","2.5","3.0","3.5","4.0","4.5","5.0","5.5","6.0"])
```
